```python
def plot_lake(name,sf):
    lago = []
    for i in range(len(sf)):
        rec = sf.shapeRecord(i)
        if name in rec.record['NAME'].lower():
            lago.append(rec)
    all_points = []
    for shapeRec in lago:
        all_points.extend(shapeRec.shape.points)
    import matplotlib.pyplot as plt
    x = []
    y = []
    for point in all_points:
        x.append(point[0])
        y.append(point[1])
    plt.scatter(x,y)
plot_lake('lugano',sf)
```
